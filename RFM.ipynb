```python
df["InvoiceDate"].max()
```

```python
dateT=dt.datetime(2011,12,11)
```

```python
RFM =df.groupby("CustomerID").agg({"InvoiceDate":lambda date:(dateT - date.max()).days,
                                  'InvoiceNo': lambda num : num.nunique(),
                                  'Total price': lambda Totalprice : Totalprice.sum()})
RFM
```

```python
RFM.columns=["Recancy","Frequency","Monetary"]
RFM
```

```python
RFM=RFM[RFM['Monetary']>0]
RFM
```

```python
RFM.describe().T
```

```python
RFM['Recancy_score']=pd.qcut(RFM['Recancy'], 5, labels=[5,4,3,2,1])
```

```python
RFM['Frequency_score']=pd.qcut(RFM['Frequency'].rank(method='first'), 5 ,labels=[1,2,3,4,5])
```

```python
RFM['Monetary_score']=pd.qcut(RFM['Monetary'], 5 ,labels=[1,2,3,4,5])
```

```python
RFM.head()
```

```python
RFM['RFM_Score']=(RFM['Recancy_score'].astype(str) + RFM['Frequency_score'].astype(str)+ RFM['Monetary_score'].astype(str))
```
